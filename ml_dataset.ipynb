```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier


crop_data = pd.read_csv("Crop.csv")
production_data = pd.read_csv("Crop_recommendation.csv")

X = crop_data[['N','P','K','ph']]
y = crop_data['label']

model = RandomForestClassifier()
model.fit(X,y)


prediction = model.predict([[90,40,40,6.5]])
predicted_crop = prediction[0]

print("🌱 Predicted Crop:", predicted_crop)




print("\nProduction Columns:", production_data.columns)

# 🔴 CHANGE 'Commodity' based on your dataset
column_name = 'Commodity'   # change if needed

if column_name in production_data.columns:
    filtered = production_data[
        production_data[column_name].str.lower() == predicted_crop.lower()
    ]

    print("\n📊 Production Data for", predicted_crop)
    print(filtered.head())
else:
    print("\n⚠ Column not found. Check production dataset column name.")
```
